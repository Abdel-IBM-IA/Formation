### Introduction:

Ce notebook est basé sur les données du titanic que l'on trouve sur [Kaggle](https://www.kaggle.com/garikhakobyan/titanic-full-data) ou sur le drive titanic3.csv


###  1. Les imports

```python
# Les imports et reglages classiques
import pandas as pd
import numpy as np

# Quelques réglages
pd.set_option('display.notebook_repr_html', False)
pd.set_option('display.max_columns', 20)
pd.set_option('display.max_rows', 25)

%matplotlib inline
import matplotlib as mpl
import matplotlib.pyplot as plt
```

### 2. Créer un dataframe *titanic* à partir des données fournies dans le fichier *titanic.csv* et indexer par PassengerId

```python
titanic = pd.read_csv('titanic.csv', sep = ',')
titanic = titanic.set_index('PassengerId')
titanic.head()
```

### 3. Créer un diagramme circulaire représentant la répartition hommes/femmes

```python
PassengerByGender = titanic.Sex.value_counts(normalize=True)*100

MaleProportion = PassengerByGender[0]
FemaleProportion =  PassengerByGender[1]

labels = 'Hommes', 'Femmes'
sizes = [MaleProportion, FemaleProportion]
colors = ['lightskyblue', 'lightcoral']
explode = (0.1, 0)

plt.pie(sizes, explode=explode, labels=labels, colors=colors, 
        autopct='%1.1f%%', shadow=True, startangle=90)

plt.axis('equal')

plt.savefig('PassengerByGender.png')
plt.show()
```

### 4. Créer un nuage de points représentant les variations de la variable Fare avec l'age, différencier la couleur des points selon le sexe.

```python
titanic['color']=titanic['Sex'].apply(lambda x : 'Blue' if x=='male' else 'red')

FareByAgeAndSex = titanic.plot.scatter('Age','Fare',c='color')
FareByAgeAndSex.set(title = 'Fare / Age')
```

### 5. Créer un histogramme représentant les tarifs des billets

```python
plt.hist(titanic.Fare, bins = 'auto')

#Legende
plt.xlabel('Fare')
plt.ylabel('Frequence')
plt.title('Repartition des prix des billets')

plt.show()
```
