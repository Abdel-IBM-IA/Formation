### 1. Importer les librairies nécessaires

```python
import pandas as pd
import numpy as np
```

### 2. Recharger le jeu de données chipoltle.csv dans un dataFrame nommé chipo

```python
data = pd.read_csv('data/chipotle.tsv', sep='\t')
data
```

###  3. Combien de produits coûtent plus de $10.00?

```python
data.fillna('')
data.item_price = data.item_price.apply(lambda x: float(x[1:]))
data.head()
```

```python
dataInf10 = data[data.item_price >= 10]

dataInf10UniqueItem = dataInf10['item_name'].unique()
len(dataInf10UniqueItem)
```

###  4. Quel est le prix de chaque produit ? 
###### Afficher un dataframe de 2 colonnes item_name et item_price

```python
data['item_price_unit'] = data['item_price']/data['quantity']
dataNamePrice = ['item_name' , 'item_price_unit']
dataNamePriceUnique = data[dataNamePrice].drop_duplicates()
dataNamePriceUnique
```

### 5. Trier nom d'article

```python
sorted(dataNamePriceUnique['item_name'].unique())
```

### 6. Quelle était la quantité commandée de l'article le plus cher ?

```python
dataPriceUnitMax = dataNamePriceUnique.sort_values(['item_price_unit'], ascending=False)
articleMoreExpensive = dataPriceUnitMax.iloc[0][0]

data[data.item_name==articleMoreExpensive].quantity.sum()
```

### 7. Combien de fois l'article Veggie Salad Bowl a t-il été commandé?

```python
len(data[data.item_name=='Veggie Salad Bowl'])
```
